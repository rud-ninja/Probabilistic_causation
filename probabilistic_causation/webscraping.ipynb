```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
```

```python
base_url = 'https://en.soccerwiki.org'
url = 'https://en.soccerwiki.org/search/player?firstname=&surname=&nationality=&leagueid=36&position=&minrating=60&maxrating=99&minage=15&maxage=60&country=&minheight=150&maxheight=220&foot=&submit='
```

```python
temp = 1
data = []

while True:
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    player_table = soup.find('table', class_ = 'table-custom table-roster')
    
    for player in player_table.find_all('tbody'):
        rows = player.find_all('tr')
        for row in rows:
            player_name = row.find_all('td', class_ = 'text-left')[0].text
            team = row.find_all('td', class_ = 'text-left')[1].text
            pos = row.find('td', class_ = 'text-left text-dark').text
            height = row.find_all('td', class_ = 'text-center text-dark')[0].text
            foot = row.find_all('td', class_ = 'text-center text-dark')[1].text
            age = row.find_all('td', class_ = 'text-center text-dark')[2].text
            rating = row.find_all('td', class_ = 'text-center text-dark')[3].text
            data.append((player_name, team, pos, height, foot, age, rating))
    
    print("No of players", len(data))
    next_page = soup.find_all('li', class_ = 'page-item')

    if len(next_page)==2:
        next_url = next_page[1].find('a', class_ = 'page-link').get('href')
        url = urljoin(base_url, next_url)

        
    elif len(next_page)==1 and temp==1:
        next_url = next_page[0].find('a', class_ = 'page-link').get('href')
        url = urljoin(base_url, next_url)
        temp=0
        
        
    else:
        break
        
```

```python
import pandas as pd

df = pd.DataFrame.from_records(data, columns =['Name', 'Team', 'Position', 'Height', 'Strong foot', 'Age', 'Rating'])
df
```

```python
df.to_csv('Ligue1_players.csv', encoding='utf-8', index=False)
```
